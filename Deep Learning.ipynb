# Load Data

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = loadtxt("pima-indians-diabetes.csv", delimiter=",")
# Split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(8,), activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

# Compile keras model

The model can be compiled so that it can use the backend, which will be Theano or TensorFlow. This then chooses the best way to represent the network for training and making predictions, based on user hardware.

To compile you need to specify a loss function: https://machinelearningmastery.com/how-to-choose-loss-functions-when-training-deep-learning-neural-networks/

You also need an optimizer. Here we use the Adam Optimization Engine: https://machinelearningmastery.com/adam-optimization-algorithm-for-deep-learning/

We also collect and report the classification accuracy since this is a classification problem.

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# That was lightning fast...
```

# Fit keras model. 

Training occurs over epochs (one pass through all rows in the training dataset) and each is split into batches, which can be one or more samples and is thae amount considered by the model before the weights are updated.

This one uses 150 epochs and a batch size of 10. 

Configurations can be chosen experimentally by trial and error. You want to look for model convergence: that is, a good mapping of input to prediction without overusing resource. 

```python
model.fit(X,y, epochs=150, batch_size=10)
```

```python
# Evaluate accuracy

_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
